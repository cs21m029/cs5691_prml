```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVC

from google.colab import drive
drive.mount('/content/drive')
```

```python
data = np.load('drive/My Drive/Sem1_Aug-Nov_2021/PRML_CS5691/A2/3/data/Dataset_Kernel_Train.npy')
data1 = np.load('drive/My Drive/Sem1_Aug-Nov_2021/PRML_CS5691/A2/3/data/Dataset_Kernel_Test.npy')
```

```python
#Train data
#print(data[0])
#print(data.shape)

data_train = data.T
X_ = data_train[0:2][:]
X_train = X_.T
X1_train = data_train[0][:]
X2_train = data_train[1][:]
Y_train = data_train[2][:]

#print(X_train.shape)
#print(X1_train.shape)
#print(X2_train.shape)
#print(Y_train.shape)
#print(np.unique(Y_train))

#Test data
#print(data[0])
#print(data1.shape)

data_test = data1.T
X1_ = data_test[0:2][:]
X_test = X1_.T
X1_test = data_test[0][:]
X2_test = data_test[1][:]
Y_test = data_test[2][:]


#print(X_test.shape)
#print(X1_test.shape)
#print(X2_test.shape)
#print(Y_test.shape)
#print(np.unique(Y_test))
```

```python
# plotting Training data, Q3 (a)

plt.figure(figsize=(15,8))
plt.scatter(X1_train,X2_train,s=9,c=Y_train)
plt.xlabel('X1',fontsize=14)
plt.ylabel('X2',fontsize=14)
plt.show()
```

```python
#  Q3 (b)

# creating svc object
svm_c = SVC(C = 1 , degree=3 , gamma=0.5 ,kernel = 'rbf' )

# training svc using training data
svm_c = svm_c.fit(X_train,Y_train)

# making prediction for test data
Y_pred = svm_c.predict(X_test)

# calculating accuracy for test data
print (f'Accuracy : {svm_c.score(X_test,Y_test):.3f}')

# plotting output
plt.figure(figsize=(15,8))
plt.scatter(X1_test,X2_test,s=9,c=Y_pred)
plt.xlabel('X1',fontsize=14)
plt.ylabel('X2',fontsize=14)
plt.show()
```
